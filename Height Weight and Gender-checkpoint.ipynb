# Request
Create the data frame rejects including the following: Cases with a desired bmi < 10 Cases with a desired bmi > 50 and a desired to actual ratio > 1 Remove all cases in the dataframe rejects from the dataframe data.

```python
import pandas as pd

# Load the CSV file
file_path = 'cdc.csv'
df = pd.read_csv(file_path)

# Calculate BMI and add it to the dataframe
df['bmi'] = (df['weight'] / (df['height'] ** 2)) * 703

# Create a categorical version of BMI
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25.0 <= bmi < 29.9:
        return 'Overweight'
    elif 30.0 <= bmi < 39.9:
        return 'Obese'
    else:
        return 'Morbidly obese'

df['bmi_cat'] = df['bmi'].apply(categorize_bmi)

# Calculate desired BMI (bmi_des) and add it to the dataframe
df['bmi_des'] = (df['wtdesire'] / (df['height'] ** 2)) * 703

# Create a categorical version of desired BMI
def categorize_bmi_des(bmi_des):
    if bmi_des < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi_des < 24.9:
        return 'Normal weight'
    elif 25.0 <= bmi_des < 29.9:
        return 'Overweight'
    elif 30.0 <= bmi_des < 39.9:
        return 'Obese'
    else:
        return 'Morbidly obese'

df['bmi_des_cat'] = df['bmi_des'].apply(categorize_bmi_des)

# Calculate DesActRatio and add it to the dataframe
df['DesActRatio'] = df['bmi_des'] / df['bmi']

# Create the dataframe rejects
rejects = df[(df['bmi_des'] < 10) | ((df['bmi_des'] > 50) & (df['DesActRatio'] > 1))]

# Remove all cases in the dataframe rejects from the original dataframe
df_cleaned = df.drop(rejects.index)

# Display the first few rows of the cleaned dataframe and the rejects dataframe
df_cleaned.head(), rejects.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_cleaned is your dataframe

# Ensure the columns are present
required_columns = ['bmi', 'bmi_des', 'DesActRatio', 'gender', 'age']
missing_columns = [col for col in required_columns if col not in df_cleaned.columns]

if missing_columns:
    raise ValueError(f"The dataframe does not contain the required columns: {', '.join(missing_columns)}")

# Drop rows with NaN values in the relevant columns
df_cleaned = df_cleaned.dropna(subset=required_columns)

# Create age categories
age_bins = [18, 32, 44, 58, 100]
age_labels = ['18-31', '32-43', '44-57', '58-99']
df_cleaned['ageCat'] = pd.cut(df_cleaned['age'], bins=age_bins, labels=age_labels, right=False)

# Drop rows with NaN values in the ageCat column
df_cleaned = df_cleaned.dropna(subset=['ageCat'])

# Create the scatter plot with loess smoother and a red abline faceted by gender and age category
g = sns.FacetGrid(df_cleaned, row="gender", col="ageCat", height=4, aspect=1.2)
g.map_dataframe(sns.scatterplot, x='bmi', y='bmi_des', hue='DesActRatio', palette='Blues', alpha=0.1)

# Add loess smoother and red abline to each facet
for ax in g.axes.flatten():
    gender = ax.get_title().split(' | ')[0].split('=')[1].strip()
    ageCat = ax.get_title().split(' | ')[1].split('=')[1].strip()
    sns.regplot(x='bmi', y='bmi_des', data=df_cleaned[(df_cleaned['gender'] == gender) & (df_cleaned['ageCat'] == ageCat)], scatter=False, lowess=True, ax=ax, line_kws={'color': 'yellow', 'alpha': 0.5})
    ax.axline((0, 0), slope=1, color="red")
    ax.set_xlim(10, df_cleaned['bmi'].max())
    ax.set_ylim(7, df_cleaned['bmi_des'].max())
    ax.grid(True)

# Customize the color legend for DesActRatio with specific values 0.5, 1.0, and 1.5
norm = plt.Normalize(0.5, 1.5)
sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
sm.set_array([])
cbar = g.fig.colorbar(sm, ax=g.axes, orientation="vertical", pad=0.02, aspect=50)
cbar.set_label("DesActRatio")
cbar.set_ticks([0.5, 1.0, 1.5])

# Add alpha legend
handles, labels = g.axes[0, 0].get_legend_handles_labels()
alpha_legend = plt.Line2D([0], [0], marker='o', color='w', label='Alpha 0.1', markerfacecolor='blue', markersize=10, alpha=0.1)
handles.append(alpha_legend)
g.fig.legend(handles=handles, title='DesActRatio', loc='center right', bbox_to_anchor=(1.15, 0.5))

g.set_titles("{col_name} | {row_name}")
plt.subplots_adjust(right=0.85, top=0.9)
g.fig.suptitle('Actual vs Desired BMI by Gender and Age', size=16)

plt.show()
```
